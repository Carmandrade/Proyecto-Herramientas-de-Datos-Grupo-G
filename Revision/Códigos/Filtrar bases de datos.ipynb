```python
#Funcion para calcular promedio ponderado de las tasas de mortalidad para una region
def calcular_promedio_ponderado(datos_tasas, datos_poblacion):
    datos_poblacion = datos_poblacion.drop(datos_poblacion.columns[:3], axis=1)
    datos_tasas = datos_tasas.drop(datos_tasas.columns[:4], axis=1)
    suma_columnas_1 = datos_poblacion.sum().tolist()#Sumas totales publación
    datos_m = datos_poblacion*datos_tasas
    suma_columnas_2 = datos_m.sum().tolist()#Multiplicacion de tasas por poblacion de cada pais a cada año respectivo
    promedios = [a / b for a, b in zip(suma_columnas_2, suma_columnas_1)]#Calcula promedio ponderado la tasa de mortalidad
    
    return (promedios)
```

```python
#Funcion que calcula el PIB total de una region
def calcular_pib_total_por_sexo(datos_pib,datos_poblacion):
    datos_poblacion = datos_poblacion.drop(datos_poblacion.columns[:3], axis=1)
    datos_pib = datos_pib.drop(datos_pib.columns[:4], axis=1)
    datos_m = datos_poblacion*datos_pib
    suma_columnas_1 = datos_poblacion.sum().tolist()#Sumas totales publación
    suma_columnas_2 = datos_m.sum().tolist()#Multiplicacion de tasas por poblacion de cada pais a cada año respectivo
    promedios = [a / b for a, b in zip(suma_columnas_2, suma_columnas_1)]#Calcula promedio ponderado la tasa de mortalidad
    return (promedios)
```

```python
#Formula que permite calcular una regresion lineal entre tasas de mortalidades y pib per capita
def regresionLineal(dt_mortalidad, dt_PIB):
    datos = pd.DataFrame({'Mortalidad': dt_mortalidad, 'PIB': dt_PIB})
    y = datos['Mortalidad']
    x = datos['PIB']
    mod = smf.ols('y ~ x', data=datos).fit()
    return mod
```
